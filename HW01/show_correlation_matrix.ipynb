```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
data = pd.read_csv('./covid_train.csv')
print(data.shape)
print(data.corr().shape)
feats = []
for i,corr in enumerate(data.corr()['tested_positive.2'].abs()):
    if corr > 0.6:
        feats.append(i)
print(feats)
```

```python
plt.matshow(data.corr())
plt.colorbar()
plt.title('Correlation Matrix', fontsize=16)
# cb.ax.tick_params(labelsize=14)
```
